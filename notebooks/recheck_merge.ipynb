```python
from bs4 import BeautifulSoup
import feedparser
from datetime import date
import pandas as pd
import numpy as np
import re  
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import time

stop_words = stopwords.words('english')
```

```python
df = pd.read_csv('newspaper/static/datasets/ht.csv')
df1 = pd.read_csv('newspaper/static/datasets/kt.csv')
df2 = pd.read_csv('newspaper/static/datasets/lk.csv')
df3 = pd.read_csv('newspaper/static/datasets/ns.csv')
df4 = pd.read_csv('newspaper/static/datasets/nt.csv')
df5 = pd.read_csv('newspaper/static/datasets/ok.csv')
df6 = pd.read_csv('newspaper/static/datasets/tn.csv')
```

```python
df.drop('category', inplace=True, axis=1)
df1.drop('category', inplace=True, axis=1)
df2.drop('category', inplace=True, axis=1)
df3.drop('category', inplace=True, axis=1)
df4.drop('category', inplace=True, axis=1)
df5.drop('category', inplace=True, axis=1)
df6.drop('category', inplace=True, axis=1)
```

```python
df['newspaper'] = 'ht'
df1['newspaper'] = 'kt'
df2['newspaper'] = 'lk'
df3['newspaper'] = 'ns'
df4['newspaper'] = 'nt'
df5['newspaper'] = 'ok'
df6['newspaper'] = 'tn'
```

```python
df.shape[0] + df1.shape[0] + df2.shape[0] + df3.shape[0] + df4.shape[0] + df5.shape[0] + df6.shape[0]
```

```python
temp = pd.concat([df, df1, df2, df3, df4, df5, df6], sort=False).dropna().reset_index(drop=True)
```

```python
def cleanHTML(raw_html):
    text = BeautifulSoup(raw_html, "lxml").text
    word_tokens = word_tokenize(text.lower().rstrip()) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    return ' '.join(filtered_sentence)
```

```python
temp
```

```python
temp['content'] = temp.content.apply(cleanHTML)
```
